```python
import sqlite3
import pandas as pd

# Connect to the database
db_path = "../data/clean/finances.db"
conn = sqlite3.connect(db_path)

# Read expenses table
expenses = pd.read_sql_query("SELECT * FROM expenses", conn)
print("Expenses table:")
print(expenses.head())

# Read incomes table
incomes = pd.read_sql_query("SELECT * FROM incomes", conn)
print("\nIncomes table:")
print(incomes.head())

# Close the connection
conn.close()
```
